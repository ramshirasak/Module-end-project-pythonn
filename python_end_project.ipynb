5.FIND IF THERE IS ANY CORRELATION BETWEEN AGE AND SALARY , REPRESENT IT VISUALLY.

```python
correlation = df['Salary'].corr(df['Age'])
print("THE CORRELATION B/w Salary AND Age IS:",correlation)
```

```python
#Graphical representation of correlation between afe and salary
```

```python
sns.scatterplot(x="Age", y= "Salary", data= df, edgecolor='b')
plt.xlabel("Age",color='y')
plt.ylabel("Salary",color='y')
plt.title("CORRELATION BETWEEN SALARY & AGE",color='r')
plt.show()
```
